## Derivatives with single variable

```python
f(x) = a*x^5 + b*x^3 + c*x
f.show()
```

```python
df(x)=f.derivative(x)
df.show()
```

```python
diff(e^sin(x^2),x).show()
```

```python
diff(arctan(x),x).show()
```

## Visualizing functions and derivatives

```python
g(x) = x^3 - x + 2
plot(g(x),-2,2,color='blue')
```

We can take a derivative of the function and superpose the plots to understand the variations in the slope.

```python
gprime(x) = g(x).derivative()
fig1 = plot(g(x),-2,2,color='blue')
fig2 = plot(gprime(x),-2,2,color='red')
fig = fig1 + fig2
fig.show()
```

## Partial derivatives of multivariable functions

```python
p(x,y) = 3*x^4*y^3 + 9*y*x^2 - 4*x + 8*y
p.show()
```

Here we evaluate $ {\partial p \over \partial x} $

```python
dpdx(x,y) = diff(p,x)
dpdx.show()
```

Here we evaluate $ {\partial p \over \partial y} $

```python
dpdy(x,y)=diff(p,y)
dpdy.show()
```

One can check if the order of differentiation matters.

Here we evaluate $ {\partial^2 p \over \partial x \partial y} $

```python
d2pdxy(x,y)=diff(p,x,y)
d2pdxy.show()
```

Here we evaluate $ {\partial^2 p \over \partial y \partial x} $

```python
d2pdyx(x,y)=diff(p,y,x)
d2pdyx.show()
```

## Double differentiation

Here we evaluate $ {\partial^2 p \over \partial x^2} $

```python
d2pdx2(x,y)=diff(p,x,2)
d2pdx2.show()
```

## Implicit differentiation

In situations when a function is not defined explicity but we need to evaluate its derivative, implicit differentiation is useful. For example:

Given
$$ 5 h(x)^2 + \sin(h) = x^2 $$
Evaluate $ {\partial h \over \partial x}$

```python
h(x) = function('h')(x)
expr = 5*h^2 + sin(h) == x^2
expr.show()
```

```python
dhdx = solve(diff(expr),diff(h))
dhdx
```
